#Prepare data

```python
# Import library
import spacy
import pandas as pd
from spacy.matcher import Matcher
```

```python
# Open the text from cloud drive. If you store it at local drive, remember to change the open path
with open("/content/drive/MyDrive/TextExtraction/result.txt", "r") as f:
    text = str(f.read())
```

```python
# Loading model for tokennization
nlp = spacy.load("en_core_web_sm")
```

#1.Get the RAW information of BLOCK

```python
# Get the BLOCK_INFO
matcher = Matcher(nlp.vocab)
# Build the pattern to extract block information
pattern = [{"LOWER": "block", "OP": "+"},
           {"POS" : "PUNCT", "OP" : "?"},
           {"POS" : "SPACE", "OP" : "?"},
           {"POS" : "NUM", "OP" : "+"}]
matcher.add("PROPER_NOUNS", [pattern])
doc = nlp(str(text))
matches = matcher(doc)
print (len(matches))

Block_list = []
Block_string = ""
for match in matches:
  print(str((doc[match[1]:match[2]])))
  block_string = str((doc[match[1]:match[2]]))
  Block_list += [block_string]
print(Block_list)
df_1 = pd.DataFrame(Block_list, columns=['BLOCK_INFO'])
df_1.head(len(df_1))
```

#2.Get the RAW information of LOT

```python
#Get the LOT_INFO
matcher = Matcher(nlp.vocab)
# Build the pattern to extract lot information
pattern_Lot = [{"LOWER": "lot", "OP": "+"},
           {"POS" : "PUNCT", "OP" : "?"},
           {"POS" : "SPACE", "OP" : "?"},
           {"POS" : "NUM", "OP" : "+"}]
matcher.add("PROPER_NOUNS", [pattern_Lot])
doc = nlp(str(text))
matches = matcher(doc)
print (len(matches))

Lot_list = []
Lot_string = ""
for match in matches:
  print(str((doc[match[1]:match[2]])))
  Lot_string = str((doc[match[1]:match[2]]))
  Lot_list += [Lot_string]

print(Lot_list)
df_2 = pd.DataFrame(Lot_list, columns=['Lot_INFO'])
df_2.head(len(df_2))
```

#3.Get the RAW information of Attorney

```python
#Get Attorney_INFO
text = str(text)
nlp = spacy.load("en_core_web_sm")

phrase_matcher = PhraseMatcher(nlp.vocab)
phrases = ["Attorney", "Attorneys", "Attorney(s)", "Gross Polowy"]
patterns = [nlp(text) for text in phrases]
phrase_matcher.add('AI', None, *patterns)

doc = nlp(text)
Attorney_list = []
for sent in doc.sents:
    for match_id, start, end in phrase_matcher(nlp(sent.text)):
        if nlp.vocab.strings[match_id] in ["AI"]:
          Attorney_list += [sent.text]
        
print(Attorney_list)
df_3 = pd.DataFrame(Attorney_list, columns=['Attorney_INFO'])
df_3.head(len(df_3))
```

# 4.Get the RAW information of Foreclosure address

```python
#Get raw Foreclosure INFO
import spacy
from spacy.matcher import PhraseMatcher

text = str(text)
nlp = spacy.load("en_core_web_sm")
pattern_1 = "known as"
pattern_2 = "the premises described as follows:"
phrase_matcher = PhraseMatcher(nlp.vocab)
#phrases = ["known"]
phrases = [pattern_1, pattern_2]
patterns = [nlp(text) for text in phrases]
phrase_matcher.add('AI', None, *patterns)

doc = nlp(text)
Foreclosure_list = []
for sent in doc.sents:
    for match_id, start, end in phrase_matcher(nlp(sent.text)):
        if nlp.vocab.strings[match_id] in ["AI"]:
          Foreclosure_list += [sent.text]
        
print(Foreclosure_list)
df_4 = pd.DataFrame(Foreclosure_list, columns=['Foreclosure_property_address_RAW'])
df_4.head(len(df_4))
```

#5.Data Cleaning as much as possible
###By showing the RAW result of foreclosure address, I realize that some data are not complete, for example row 14th in Foreclosure_list. What is more, there are many unneccesarry strings in the rows. it's necerracy to clean it as much as I can 

```python
from operator import index
#Fill the missing value of 14th rows in Foreclosure_list
row_14 = Foreclosure_list[14]
print(row_14)

missing_location = text.index("Said premises known as 5200 1ST")
print("Location of the missing address in text",missing_location)

passing = text[missing_location:missing_location+53]
Foreclosure_list[14] = passing
print("Complete the 14th row information in Foreclosure_list:", passing)


#Eliminate unneccesarry infomation after Spacy
Foreclosure_list_clean = []
for rows in Foreclosure_list:
  if "known" in str(rows):
    index_1 = str(rows).index("known")
    Foreclosure_list_clean.append([str(rows)[(index_1+len(pattern_1)):]])
  elif "described" in str(rows):
    index_2 = str(rows).index("described")
    Foreclosure_list_clean.append([str(rows)[(index_2+len(pattern_2)-12):]])
  else:
    continue
print(len(Foreclosure_list_clean))
```
